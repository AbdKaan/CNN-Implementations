```python
import warnings
warnings.filterwarnings('ignore')

from keras.models import Sequential
from keras.layers import Conv2D, Dense, Activation, MaxPooling2D, Dropout, Flatten, Activation
from keras.layers.normalization import BatchNormalization
from keras.optimizers import SGD
import tflearn.datasets.oxflower17 as oxflower17
from sklearn.model_selection import train_test_split
```

```python
def alexnet_model(img_shape=(228, 228, 3), classes = 1000):
    alexnet = Sequential()

    # Convolutional Layer 1
    alexnet.add(Conv2D(filters=96, kernel_size=11, strides=4, input_shape=img_shape))
    # 
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))
    alexnet.add(MaxPooling2D(pool_size=3, strides=2))

    # Convolutional Layer 2
    alexnet.add(Conv2D(filters=256, kernel_size=5, strides=1))
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))
    alexnet.add(MaxPooling2D(pool_size=3, strides=2))

    # Convolutional Layer 3
    alexnet.add(Conv2D(filters=384, kernel_size=3, strides=1))
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))

    # Convolutional Layer 4
    alexnet.add(Conv2D(filters=384, kernel_size=3, strides=1))
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))

    # Convolutional Layer 5
    alexnet.add(Conv2D(filters=256, kernel_size=3, strides=1))
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))
    alexnet.add(MaxPooling2D(pool_size=3, strides=2))

    # Dense Layer 1
    alexnet.add(Flatten())
    alexnet.add(Dense(4096))
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))
    alexnet.add(Dropout(0.5))

    # Dense Layer 2
    alexnet.add(Dense(4096))
    alexnet.add(BatchNormalization())
    alexnet.add(Activation('relu'))
    alexnet.add(Dropout(0.5))

    # Dense Layer 3
    alexnet.add(Dense(classes))
    alexnet.add(Activation('softmax'))
    
    return alexnet
```

```python
# Import dataset
X, Y = oxflower17.load_data(one_hot=True)
```

```python
print("X's shape: ", X.shape)
print("Y's shape: ", Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
print("X_train's shape: ", X_train.shape)
print("Y_train's shape: ", Y_train.shape)
```

```python
alexnet = alexnet_model((224,224,3), 17)
alexnet.summary()
```

```python
# In the paper, they initialize the learning rate by 0.01 and
# reduce it three times by dividing it by 10 prior to termination.
sgd = SGD(lr=0.01, momentum=0.9, decay=0.0005)
alexnet.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
alexnet.fit(X_train, Y_train, batch_size=128, epochs=100, validation_split=0.3, shuffle=True)
```

```python
loss_and_metrics = alexnet.evaluate(X_test, Y_test, batch_size=128)
print("Loss on test set: ", loss_and_metrics[0])
print("Accuracy on test set: ", loss_and_metrics[1])
```
